```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/home/redleaf/Documents/DUK/Mini Project/Akt.tsv', delimiter='\t')
df.head()
```

```python
df.isna().sum()
```

```python
threshold = len(df)/2
df.dropna(thresh=threshold, axis=1, inplace=True)
df.head()
```

```python
df.info()
```

```python
print("display IC50 values with '<' symbol")
display(df[df['IC50 (nM)'].astype(str).str.contains('<')])
print("Number of IC50 values with '<' symbol")
print(df['IC50 (nM)'].astype(str).str.contains('<').sum())
df['IC50 (nM)'] = df['IC50 (nM)'].str.replace('<','')
print("after removing")
print("Number of IC50 values with '<' symbol")
print(df['IC50 (nM)'].astype(str).str.contains('<').sum())
print("display IC50 values with '<' symbol")
display(df[df['IC50 (nM)'].astype(str).str.contains('<')])
```

```python
print("display IC50 values with '>' symbol")
display(df[df['IC50 (nM)'].astype(str).str.contains('>')])
print("Number of IC50 values with '>' symbol")
print(df['IC50 (nM)'].astype(str).str.contains('>').sum())
df['IC50 (nM)'] = df['IC50 (nM)'].str.replace('>','')
print("after removing")
print("Number of IC50 values with '>' symbol")
print(df['IC50 (nM)'].astype(str).str.contains('>').sum())
print("display IC50 values with '>' symbol")
display(df[df['IC50 (nM)'].astype(str).str.contains('>')])
```

```python
display(df[df['IC50 (nM)'].astype(str).str.contains('\+')])
print(df['IC50 (nM)'].astype(str).str.contains('\+').sum())
```

```python
df["IC50 (nM)"] = df["IC50 (nM)"].apply(pd.to_numeric)
df["BindingDB Reactant_set_id"] = df["BindingDB Reactant_set_id"].apply(pd.to_numeric)
```

```python
df.info()
```

```python
MIN = df['IC50 (nM)'].min()
print("minimum of IC50 (nM) :",MIN)
MAX = df['IC50 (nM)'].max()
print("minimum of IC50 (nM) :",MAX)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Ligand SMILES', y='IC50 (nM)', data=df)
plt.xlabel('Ligand SMILES')
plt.ylabel('IC50 (nM)')
plt.title('IC50 Values for Different Ligand SMILES')
# plt.axis('off') 
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
# plt.yticks(np.linspace(0,500000,500))
plt.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)
plt.tight_layout()  # Adjust layout to prevent overlapping labels
plt.show()
```
